```python
# (A29) Write a Python function to get the largest number, smallest num and sum of all from a list

def find_largest_smallest_sum(numbers):
  """
  This function takes a list of numbers and returns the largest number,
  smallest number, and the sum of all numbers in the list.
  """
  if not numbers:
    return None, None, 0

  largest = numbers[0]
  smallest = numbers[0]
  total = 0

  for number in numbers:
    if number > largest:
      largest = number
    if number < smallest:
      smallest = number
    total += number

  return largest, smallest, total
```

```python
# (A31) Write a Python program to count the number of strings where the string length is 2 or more and the first and last character are same from a given list of strings.

def count_strings(strings):
  """
  This function counts the number of strings where the string length is 2 or more
  and the first and last character are the same from a given list of strings.
  """
  count = 0
  for string in strings:
    if len(string) >= 2 and string[0] == string[-1]:
      count += 1
  return count

# Example
strings = ["abc", "xyz", "aba", "1221"]
result = count_strings(strings)
print("Number of strings:", result)
```

```python
 # (A32) Write a Python program to remove duplicates from a list.

def remove_duplicates(lst):
  """
  This function removes duplicates from a list.
  """
  unique_lst = []
  for item in lst:
    if item not in unique_lst:
      unique_lst.append(item)
  return unique_lst

# Example
my_list = [1, 2, 2, 3, 4, 4, 5]
result = remove_duplicates(my_list)
print("List after removing duplicates:", result)
```

```python
# (A39) Write a Python program to find the second smallest number in a list.

def find_second_smallest(numbers):
  if len(numbers) < 2:
    return None

  smallest = min(numbers[0], numbers[1])
  second_smallest = max(numbers[0], numbers[1])

  for i in range(2, len(numbers)):
    if numbers[i] < smallest:
      second_smallest = smallest
      smallest = numbers[i]
    elif numbers[i] < second_smallest and numbers[i] != smallest:
      second_smallest = numbers[i]

  return second_smallest
```

```python
# (A43) What is tuple? Difference between list and tuple.
#A tuple is an ordered, immutable collection of items.
# It is similar to a list, but the key difference is that tuples cannot be modified once created.

# Differences:

# Lists:
# - Mutable: You can add, remove, or modify elements after creation.
# - Defined using square brackets: []
# - More flexible for dynamic data.

# Tuples:
# - Immutable: Once created, you cannot change the elements.
# - Defined using parentheses: ()
# - Suitable for data that should not be altered.

# Example of a list:
my_list = [1, 2, 3]
my_list[0] = 4  # Modifying the list
print(my_list)  # Output: [4, 2, 3]

# Example of a tuple:
my_tuple = (1, 2, 3)
# my_tuple[0] = 4  # This would cause an error because tuples are immutable
print(my_tuple)  # Output: (1, 2, 3)
```

```python
'''(59) Write a Python program to create a dictionary from a string.
Note: Track the count of the letters from the string.'''

def count_letters(input_string):
    letter_count = {}

    for char in input_string:
        char = char.lower()

        if char.isalpha():
            if char in letter_count:
                letter_count[char] += 1
            else:
                letter_count[char] = 1

    return letter_count

    input_string = input("Enter a string: ")
    result = count_letters(input_string)
    print("Letter count:", result)
```

```python
'''(63) Write a Python function to check whether a number is perfect or not. '''

def is_perfect(number):
    if number <= 0:
        return False
    sum_of_divisors = sum(i for i in range(1, number) if number % i == 0)
    return sum_of_divisors == number
```

```python
print(is_perfect(-28))
print(is_perfect(28))
```

```python
'''(74) Write a Python program to read first n lines of a file.'''

def read_first_n_lines(filename, n):
    try:
        with open(filename, 'r') as file:
            lines = [next(file) for _ in range(n)]
        return ''.join(lines)
    except FileNotFoundError:
        return "The file was not found."
    except StopIteration:
        return "The file has fewer than {} lines.".format(n)
```

```python
'''(75) Write a Python program to read last n lines of a file. '''

def read_last_n_lines(filename, n):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
        last_n_lines = lines[-n:] if n <= len(lines) else lines
        return ''.join(last_n_lines)
    except FileNotFoundError:
        return "The file was not found."
```

```python
'''(78) Write a python program to find the longest words.'''

def find_longest_words(text):
    words = text.split()
    if not words:
        return "The text is empty."
    max_length = len(max(words, key=len))
    longest_words = [word for word in words if len(word) == max_length]

    return longest_words
    text = "This is a sample text to find the longest words."
    longest_words = find_longest_words(text)
    print("Longest words:", longest_words)
```

```python
'''(80) Write a Python program to count the frequency of words in a file.'''

from collections import Counter
import re

def count_word_frequencies(filename):
    try:
        with open(filename, 'r') as file:
            text = file.read()
        words = re.findall(r'\b\w+\b', text.lower())
        word_count = Counter(words)
        return word_count
    except FileNotFoundError:
        return "The file was not found."
```

```python
filename = ''  # Replace with your file name
word_frequencies = count_word_frequencies(filename)
print(word_frequencies)
```

```python
'''(81) Write a Python program to write a list to a file.
'''

def write_list_to_file(filename, data_list):
    try:
        with open(filename, 'w') as file:
            for item in data_list:
                file.write(f"{item}\n")
        print(f"Data successfully written to {filename}.")
    except IOError as e:
        print(f"An error occurred: {e}")
```

```python
filename = 'output.txt'  # Replace with your file name
data_list = ['apple', 'banana', 'cherry', 'date']
write_list_to_file(filename, data_list)
print(filename)
```

```python
'''(82) Write a Python program to copy the contents of a file to another file.'''

def copy_file_contents(source_filename, destination_filename):
    try:
        with open(source_filename, 'r') as source_file:
            content = source_file.read()

        with open(destination_filename, 'w') as destination_file:
            destination_file.write(content)

        print(f"Contents copied from {source_filename} to {destination_filename}.")
    except FileNotFoundError:
        print("Source file not found.")
    except IOError as e:
        print(f"An error occurred: {e}")
```

```python
source_filename = ''      # Replace with your source file name
destination_filename = ''  # Replace with your destination file name
copy_file_contents(source_filename, destination_filename)
```

```python
'''(90) Write python program that user to enter only odd numbers, else
will raise an exception.'''

def get_odd_number():
    while True:
        try:
            user_input = input("Enter an odd number: ")
            number = int(user_input)
            if number % 2 == 0:
                raise ValueError("The number is not odd. Please enter an odd number.")
            print(f"Thank you! You entered the odd number: {number}")
            return number
        except ValueError as e:
            print(e)
            print("Please try again.")
```
